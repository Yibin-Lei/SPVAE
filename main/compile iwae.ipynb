```python
import glob
import pickle
import matplotlib.pyplot as plt
import os
```

```python
import pandas as pd
import numpy as np
```

```python
def compile_attributes(model="iwae", dataset="mnist_perturbed"):
    count = 0
    all_attributes = []
    for filename in sorted(glob.glob(os.path.join(model, dataset, "-1", "*", "attributes.pk"))):
        with open(filename, "rb") as f:
            attributes = pickle.load(f)
            all_attributes.append(attributes)
    for i, fn in enumerate(sorted(glob.glob(os.path.join(model, dataset, "-1", "*", "attributes.pk")))):
        print(i, fn)
    print("Found {} attributes".format(len(all_attributes)))
    cols = ['nz',  'nnodes_recog', 'nnodes_gener', 'num_weights', 'wall_time', 'test_ll', 'test_curve', 'max_validation']
    df = pd.DataFrame(columns=cols)
    for i, att in enumerate(all_attributes):
#         print(len(att["test_curve"]) - np.argmax(att["test_curve"]))
        att["test_curve"] = att["test_curve"][-12]
        if att["test_curve"] > att["test_ll"]:
            count+=1
        row = [att[c] for c in cols]
        df.loc[i] = row
    print(count)
    return df
```

```python
df = compile_attributes("iwae", "mnist_perturbed")
df.sort_values("max_validation", ascending=False)
```
