```python
import glob
import pickle
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
import os
```

```python
import pandas as pd
import numpy as np
```

```python
def compile_attributes(model="iwae", dataset="mnist_perturbed"):

    all_attributes = []
    for filename in sorted(glob.glob(os.path.join(model, dataset, "-1", "*", "attributes.pk"))):
        with open(filename, "rb") as f:
            attributes = pickle.load(f)
            all_attributes.append(attributes)
    for i, fn in enumerate(sorted(glob.glob(os.path.join(model, dataset, "-1", "*", "attributes.pk")))):
        print(i, fn)
    print("Found {} attributes for {}".format(len(all_attributes), model))
    cols = ['nz',  'nnodes_recog', 'nnodes_gener', 'num_weights', 'take_frac', 'wall_time', 'test_ll', 'test_curve', 'max_validation']
    df = pd.DataFrame(columns=cols)
    for i, att in enumerate(all_attributes):
#         print(len(att["test_curve"]) - np.argmax(att["test_curve"]))
        att["test_curve"] = att["test_curve"][-12]
        row = [att[c] for c in cols]
        df.loc[i] = row
    df = df.sort_values("take_frac", ascending=False)
    out = (df["take_frac"], df["test_ll"])
    return out
```

```python
dataset = "mnist_perturbed"
models = ("iwae","sp_iwae")
for model in models:
    (x,y) = compile_attributes(model, dataset)
    plt.plot(x,y, label=model)
plt.xlabel("Fraction of Training Set")
plt.ylabel("5 Sample ELBO on Test Set")
plt.title(dataset)
plt.legend(loc=4)
plt.show()
```
